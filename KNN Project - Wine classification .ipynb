```python
x = df.drop('customer_seg', axis=1)
x.head()
```

<h1 style="color: red;">Split Data set for train and testing</h1>

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=2)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

<h1 style="color: red;">KNN classification </h1>

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier(n_neighbors=5)
```

```python
knn.fit(x_train,y_train)
```

```python
accuracy = knn.score(x_test, y_test)
accuracy
```

<h1 style="color: pink;">Tuning Sensitivity of the Model: 'n_neighbors'</h1>

```python
accuracy = []
for k in range(1, 16):
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(x_train, y_train)
    accuracy.append(knn.score(x_test, y_test))
```

```python
accuracy
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(range(1, 16), accuracy, marker='o')
plt.title('Accuracy vs. n_neighbors')
plt.xlabel('n_neighbors')
plt.ylabel('Accuracy')
plt.grid(True)
plt.show()
```
